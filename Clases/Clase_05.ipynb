## Modelos Continuos Especiales
La distribución Gamma es un modelo muy popular debido a su flexibilidad. Una v.a. Gamma toma
valores en $(0, +\infty)$ y su función de densidad está dada por:
$$
f_{X}(x)=\frac{x^{\alpha-1} e^{-x / \beta}}{\Gamma(\alpha) \beta^{\alpha}}, \quad x>0
$$
donde $\alpha$ y $\beta$ son parámetros positivos, mientras que $\Gamma(\cdot)$ es la función Gamma:
$$
\Gamma(\alpha)=\int_{0}^{\infty} u^{\alpha-1} e^{-u} \mathrm{d} u
$$
que satisface las siguientes propiedades:
* $\Gamma(n)=(n-1) !, \quad n \in \mathbb{N}$
* $\Gamma(\alpha+1)=\alpha \Gamma(\alpha)$

```r
gamma(4.1)
```

Usaremos la notación $X \sim $Gamma$(\alpha, \beta)$. También debemos notar que la función de distribución acumulada para una v.a. Gamma, no tiene una representación sencilla.
### Propiedades
Sea $X\sim$Gamma$(\alpha,\beta)$, entonces:
* $\mathbb{E}(X)=\alpha \beta$
* $\mathbb{V}(X)=\alpha \beta^{2}$

Cuando $\alpha= 1$ es fijo, obtenemos en particular la distribución Exponencial. Formalmente, la v.a. $X$
sigue una distribuci´on Exponencial con parámetro $\beta$, denotado como $X \sim $exp$(\beta)$, si tiene la siguiente
función de densidad:
$$
f_{X}(x)=\frac{1}{\beta} e^{-x / \beta}
$$

## Distribución Weibull
Sea $X$ una v.a. continua tal que su función de densidad está dada por
$$
f_X(x) = \alpha\lambda^\alpha x
^{\alpha-1}
e^{
-(\lambda x)
\alpha}
, x > 0
$$
entonces diremos que $X$ sigue una distribución Weibull con parámetros $\alpha$ y $\lambda$ positivos. Usaremos la
notación $X \sim $Weibull$(\alpha, \lambda)$.


```r
hist(rweibull(100,190,0.3))
```

```r
pweibull(0.3,190,0.3)
```

```r
qweibull(0.632120558828558,190,0.3)
```
